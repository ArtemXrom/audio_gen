```python
from kokoro import KPipeline
from IPython.display import Audio, display
import soundfile as sf
import time

pipeline = KPipeline(lang_code='a')
```

```python
responses = {
    "hello": "Hello! My name is Tyomich, nice to meet you!",
}
more_responses = {
    "hello there": responses["hello"],
}
responses.update(more_responses)
```

```python
def generate_speech(text, voice='am_liam', speed=1.0):

  generator = pipeline(text, voice=voice, speed=speed)

  for gs, ps, audio in generator:
    display(Audio(data=audio, rate=24000, autoplay=True))
    time.sleep(0.1)

  return True

def get_response(user_input):
  user_input_lower = user_input.lower().strip()

  if user_input_lower in responses:
    answer = responses[user_input_lower]
    return answer
  else:
    return "Sorry, I don't understand you. Please, ask me questions from my list."
```

```python
while True:
  user_input = input("\nВАШ ВОПРОС: ")

  if user_input.lower() in ['quit', 'exit', 'stop', 'goodbye', 'bye', 'пока' ]:
    generate_speech("Goodbye! It was nice talking to you!", voice='am_liam')
    break

  response_text = get_response(user_input)

  generate_speech(response_text, voice='am_liam')
```
